```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
ativos = ['PETR4.SA','VALE3.SA','WEGE3.SA','RADL3.SA','OIBR3.SA','KNRI11.SA','SMAL11.SA','AAPL34.SA','IVVB11.SA','BTC-USD','USDBRL=X']
```

```python
inicio = '2020-03-11'
fim = '2021-08-24'
```

```python
precos = pd.DataFrame()
for i in ativos:
    precos[i] = yf.download(i, start=inicio, end=fim)['Adj Close']
```

```python
precos['BTC-BRL'] = precos['BTC-USD']*precos['USDBRL=X']
```

```python
 normalizando = precos/precos.iloc[0]      
```

```python
normalizando.plot(figsize=(8,8));
```
